# 2.Dropping irrelavent columns
 
 This steps is certainly needed in EDA because sometimes there would be column that we never use and in such cases dropping is useful.In this case,the columns such as Engine Fuel Type and Number of doors maynot be very relevant.


```
dataframe = dataframe.drop(['Market Category','Engine Fuel Type','Number of Doors'],axis=1)
dataframe.head()
```

#3. Renaming the columns


```
dataframe.rename(columns={"Engine HP": "HP","Engine Cylinders": "Cylinders","Transmission Type":"Transmission","Driven_wheels":"Driven mode","highway MPG":"MPG-H","city mpg":"MPG-C","MSRP":"price"},inplace = True)
dataframe
```

#4.checking for the duplicates,null values,missing values and removing them 

```
#finding the duplicates in the given file
duplicateDFROW = dataframe[dataframe.duplicated(keep='last')]
duplicateDFROW
```

```
#removing duplicates from the .csv file
dataframe.drop_duplicates(subset=None,keep='first',inplace=True)
dataframe.shape
```

```
#as checking for the null values is already done in 1st step now we will just try to eliminate null values
dataframe['HP'].fillna(value='median',inplace=True)
dataframe['Cylinders'].fillna(value='median',inplace=True)
```
